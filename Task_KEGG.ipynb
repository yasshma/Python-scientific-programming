```python
def parse_gene_paths(table_text):

    list_test_strs = [i for i in table_text.split('\n')]
    strs_results = []

    for j in list_test_strs:
        strs = []
        strs = [i for i in j.split('\t') if i != '']
        strs_results.append(strs)
    
    dict_GEN = {}
    for i in range(len(strs_results)-1):
        dict_GEN[strs_results[i][0]] = set(strs_results[i][2:])
        
    return dict_GEN

def count_paths_for_gene(gene_in_path, gene, path_part):
    count = 0
    for element in gene_in_path:
        if (path_part in element.split('_')) and (gene in gene_in_path[element]):
            count += 1
    return count
```
